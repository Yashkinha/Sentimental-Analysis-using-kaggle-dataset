```python
#normal=list()
#spam=list()
words=[]
for email in emails:
    f=open(email,encoding='latin-1')
    blob=f.read()
    words+=blob.split(' ')
```

```python
len(words)
```

```python
for i in range(0,len(words)):
    if not words[i].isalpha():
        words[i]=''
```

```python
words_dict=Counter(words)
```

```python
del words_dict['']
```

```python
len(words_dict)
```

```python
words_dict=words_dict.most_common(3000)
```

```python
words_dict
```

```python
features=[]
target=[]
for email in emails:
    f=open(email,encoding='latin-1')
    blob=f.read().split(' ')
    f.close()
    data=[]
    for i in words_dict:
        data.append(blob.count(i[0]))
    features.append(data)
    if 'spam' in email:
        target.append(0)
    if 'ham' in email:
        target.append(1)
```
